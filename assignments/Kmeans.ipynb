     This assignment is to give me an understanding of the kmeans algorithm in python using the Scikit learn package.  I chose to take a dataset that includes the annual spending in monetary units (m.u.) on diverse product categories in regions of Portugal.
     
     Normally, kmeans is used for unlabeled unorganized data, but for the purpose of this assignment i chose a labeled set but chose randome columns of data to try and analyze.    
 http://archive.ics.uci.edu/ml/datasets/Wholesale+customers    
       

```python
#start by importing necessary packages
import pandas as pd
import pylab as pl
from sklearn.cluster import KMeans
```

```python
#import dataset and label x and y

variables = pd.read_csv('customers.csv')

Y = variables[['Fresh']]

X = variables[['Frozen']]
```

     To estimate number of clusters, i will use an elbow graph :

```python
Nc = range(1, 20)

kmeans = [KMeans(n_clusters=i) for i in Nc]

kmeans

score = [kmeans[i].fit(Y).score(Y) for i in range(len(kmeans))]

score

pl.plot(Nc,score)

pl.xlabel('Number of Clusters')

pl.ylabel('Score')

pl.title('Elbow Curve')

pl.show()
```

 It looks as though 6 looks like a good amount as its right on the corner of the elbow..

```python
#plot and label the cluster plot

kmeans=KMeans(n_clusters=6)

kmeansoutput=kmeans.fit(Y)

kmeansoutput

pl.scatter(X,Y, c=kmeansoutput.labels_)

pl.xlabel('Fresh')

pl.ylabel('Frozen')

pl.title('6 Cluster K-Means')

pl.show()

```

  By taking a look at the plot and knowing that it seems as the most people but the same amount of frozen as fresh with a few outliers spending on mostly frozen or fresh. Also, seems as the more money is spent on frozen.
